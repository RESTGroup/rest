```python
#导入REST的PYTHON模块
import pyrest
import numpy
```

```python
#从ctrl.in文件读入计算控制信息和分子构型信息
mol = pyrest.read_ctrl("/home/igor/Documents/Package-Pool/rest_workspace/rest/examples/NH3/X3LYP/ctrl.in")
```

```python
#使用的密度泛函方法
print(" The exchange-correlation functional: %s\n" %mol.ctrl.xc.upper())
#体系的元素信息
(pos,size) = mol.geom.py_get_position()
position = numpy.array(pos).reshape(size[1],size[0]);
for x in range(4):
    print("%3s%16.8f%16.8f%16.8f" %(mol.geom.elem[x],position[x][0],position[x][1],position[x][2]))
```

```python
#做自洽场计算
scf_data = pyrest.do_scf(mol)
```

```python
#打印自洽场收敛的X3LYP的结果
print(" The X3LYP total energy: %16.8f Ha\n" % scf_data.scf_energy);
#打印密度矩阵
(dm,size) = scf_data.py_get_dm()
dm = numpy.array(dm[0]).reshape(size[1],size[0]);
for x in range(29):
    print(dm[x])
```
